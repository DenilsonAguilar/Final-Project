### The question I will be answering here is "What are the top 10 rated old movies?" 

```python
#This will be stored in a copy called df_copy
df_copy = df.copy()

#This removes non-numeric values from Year column
df_copy = df_copy[df_copy['Year'].apply(lambda x: str(x).isnumeric())]

#This converts the Year column to integer
df_copy['Year'] = df_copy['Year'].astype(int)

#This selects movies released before 1950
df_copy.loc[df_copy['Year'] < 1950, 'is_old'] = True

#This makes sure that the is_old column is boolean
df_copy['is_old'] = df_copy['is_old'].fillna(False).astype(bool)

#This sort the movies by rating to get the top 10 highest rated old movies
top_10_old_movies = df_copy[df_copy['is_old']].sort_values('Rating', ascending=False).head(10)

#This will show the top 10 oldest movies
print("Top 10 highest rated old movies (before 1950):")
print(top_10_old_movies[['Title', 'Year', 'Rating']])

#This plot the ratings of the top 10 oldest movies in barplot form
plt.figure(figsize=(12, 8))
plt.barh(top_10_old_movies['Title'], top_10_old_movies['Rating'], color='deepskyblue')
plt.xlabel('Rating')
plt.ylabel('Movie')
plt.title('Top 10 Highest Rated Old Movies (before 1950s)')
plt.gca().invert_yaxis() #This makes it so that the top one is at the top of the barplot
plt.show()
```

###  The question I will be answering here is "Are there any trends in movie ratings based on the year of release?"

```python
#This is a copy of the DataFrame named df_copy3, I dont want to alter the original
df_copy3 = df.copy()

#This will filter out non-integer values in the 'Year' column, included because some of the column appear to have no year
df_copy3 = df_copy3[df_copy3['Year'].str.isnumeric()]

#This changes the Year column to integer and then to decade, more on the difficult side for me to figure out
df_copy3['Decade'] = (df_copy3['Year'].astype(int) // 10 * 10).astype(str) + 's'

#This calculates the average rating for each decade
decadal_avg_rating = df_copy3.groupby('Decade')['Rating'].mean()

#This is plots movie ratings trends throughout the decades
plt.figure(figsize=(20, 8))
plt.plot(decadal_avg_rating.index, decadal_avg_rating.values, marker='o')
plt.xlabel('Decade', fontsize=12)
plt.ylabel('Average Rating', fontsize=14)
plt.title('Trends in Movie Ratings Based on Decades', fontsize=10)
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout(pad=10)
plt.show()
```

It appears that the average movie rating showed quite a lot of fluctuation over the decades. In the 1920s, the average rating was around 8.13, which then dropped slightly to approximately 7.96 in the 1930s. The rating had a slight increase to 8.03 in the 1940s, followed by a peak at around 8.06 in the 1950s. However, there was a decline in the average rating in the subsequent decades, with ratings dropping to 7.97 in the 1960s, 7.96 in the 1970s, and 7.95 in the 1980s. The trend reversed slightly in the 1990s, with the average rating rising back to 7.96. However, there was a more noticeable drop in the 2000s, where the average rating reached its lowest point at 7.90. The trend shifted upwards again in the 2010s, with the average rating increasing to 7.93. The most significant increase in ratings occurred in the 2020s, where the average rating rapidly reached its peak at 9.14. These trends suggest that there have been fluctuations in movie ratings over the years, influenced by various factors such as changes in filmmaking styles, audience preferences, and critical reception. It does appear that movies nowadays are gaining higher ratings on average due to the information gathered from the recent trends.

It's worth noting that there are exceptions, as some movies with shorter runtimes can also have high ratings. ### The question I will be answering here is "Who are the top 10 directors who have the highest average ratings for their movies?"

### The question I will be answering here is "What is the distribution of ratings among different film certificates, and is there a noticeable difference in ratings between different certificates?"

```python
#This is a copy of the DataFrame named df_copy7
df_copy7 = df.copy()

#This replaces Unknown values in the Gross column with 0
df_copy7['Gross'] = df_copy7['Gross'].replace('Unknown', '0')

#This converts the Gross column into strings and removes commas and convert to float
df_copy7['Gross'] = df_copy7['Gross'].astype(str).str.replace(',', '').astype(float)

#This groups the data by certificate and calculate the average rating for each certificate
certificate_avg_rating = df_copy7.groupby('Certificate')['Rating'].mean().sort_values()

#This plots the average ratings for each certificate in a boxplot
plt.figure(figsize=(12, 6))
sns.boxplot(x='Certificate', y='Rating', data=df_copy7, palette='icefire')
plt.xlabel('Certificate')
plt.ylabel('Rating')
plt.title('Distribution of Ratings by Film Certificate')
plt.xticks(rotation=45)
plt.show()

#shows the average rating by films certificate
print("Average Ratings by Film Certificate:")
print(certificate_avg_rating)
```

The ratings seem to be evenly distributed across different certificates, indicating that there isn't a significant difference in the overall quality of films based on their certificate. The average ratings by film certificate reveal that films rated 'G' and 'A' have the highest average ratings, both at 8.0, followed closely by 'TV-14', 'Unrated', and '16' certificates, all with an average rating of 8.1. This suggests that these certificates may generally be associated with higher-quality films, as they tend to receive higher ratings from audiences and critics.
